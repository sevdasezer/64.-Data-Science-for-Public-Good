## 1) Power(Iterative):

Write a program that calculates n'th power of a given number iteratively

```python
def iter_power(x, n):
    if n==0:
        return 1
    elif n>0:
        total=1
        for i in range(n):
            total*=x
        return total
    else:
        total=1
        for i in range(abs(n)):
            total*=x
        return 1/total
    pass
```

```python
iter_power(5,-2)
```

## 2) Power(Recursive):

Write a program that calculates n'th power of a given number recursively

```python
def recur_power(x, n):
    if n==0:
        return 1
    elif n==1:
        return x
    elif n>1:
        return x * recur_power(x,n-1)
    else:
        return 1 / (x * recur_power(x,abs(n)-1))
         
```

```python
recur_power(4,-4)
```

### 3) Key with the maximum number of values:

Given dictionary d, return a key that has a maximum value associated with it

for example if d = {"a":[1,2,3], "b":[45,43,21,23], "c":[3,4,5]}

it should return "b"

```python
d = {"a":[1,2,3], "b":[45,43,21,23], "c":[3,4,5]}
```

```python
def max_key(d):
    value=0
    for e in list(d.values()):
        for i in range(len(e)):
            if e[i]>value:
                value=e[i]
    for key in list(d.keys()):
        if value in d[key]:
            print(key) 
        
```

```python
max_key(d)
```

### 4) Fibonacci serisini recursive olarak dictionary kullanarak yapmak 

fib(2) yi bir kere hesapladıysa başka bir function call'da bir daha hesaplamasın değerini dictionaryden bulup koysun gibi


```python
def fib(x):
    fib_dict={"fib(0)":1,"fib(1)":1}
    for i in range(x-1):
        fib_dict[f"fib({i+2})"]=fib_dict[f"fib({i})"]+fib_dict[f"fib({i+1})"]
    #print(fib_dict[f"fib({x})"])
    return fib_dict
        
fib(5)
```
