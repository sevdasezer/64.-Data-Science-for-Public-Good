## 1. Soru

Write a code that modifies the given matrix M (with 3 rows and 4 columns), by swapping the (whole) columns 0 and 1.

M = [[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]]

The matrix M above should become the matrix below;

M = [[12, 11, 13, 14], [22, 21, 23, 24], [32, 31, 33, 34]]

### 1. soru çözüm:

```python
M = [[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]]
for i in range(len(M)):
    M[i][0],M[i][1]=M[i][1],M[i][0] 
print(M)
```

## 2. Soru

Find the maximal element of the matrix M and print the row number and column number of the maximal element as a tuple.

M = [[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]]

The output of the code should be row number and column number of the maximal element as a tuple. (1,3)

### 2. soru çözüm:

```python
import numpy as np
M = [[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]]
arr = np.array(M)
result = np.where(arr == np.amax(arr))
coordinates = tuple(zip(result[0], result[1]))

for coord in coordinates:
    print(coord)
```

## 3. Soru

Given two lists of numbers, count how many unique numbers occur in both of them.

For example, on input

list_1 = [5, 3, 3, 10, 40]

list_2 = [40, 3, 40, 1, 2]

output must be

6

### 3. soru çözüm:

```python
list_1 = [5, 3, 3, 10, 40]
list_2 = [40, 3, 40, 1, 2]
input_=list_1+list_2
print(f"Last List :  {input_}")
output = []
for x in input_:
    if x not in output:
        output.append(x)
print(f"New List :  {output}")
print(f"Number of Elements:  {len(output)}")
```

Write a Python program that repeatedly tells the user how many tickets are left, asks for how many tickets the user wants to buy, indicates minimum and maximum numbers to buy, and reads the user's input. This continues until all tickets have been sold. At the end, the program prints how many sales transactions have been performed. For example, we might have the following interaction:

200 tickets remaining

How many do you want to buy?

Minimum 1, maximum 4

3

197 tickets remaining

How many do you want to buy?

Minimum 1, maximum 4

4

(...)

2 tickets remaining

How many do you want to buy?

Minimum 1, maximum 2

2

86 sales transactions

(You may assume that the user enters the data correctly, i.e., all values are integers between the current minimum and maximum numbers they may buy.)

### 5. soru çözüm:

```python
Tickets=200
while True:
    x=int(input("How many do you want to buy?"))
    if x<=4 and x>=1:
        Tickets-=x
        print (f"{Tickets} tickets remaining")
    else:
        print("Please, enter a value between 1 and 4")
        continue
```
